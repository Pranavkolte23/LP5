#### Linear regression by using Deep Neural network: Implement Boston housing price prediction problem by Linear regression using Deep Neural network. Use Boston House price prediction dataset. 

```python
import pandas as pd
```

```python
df = pd.read_csv('HousingData.csv')
df.head()
```

```python
df.isna().sum()
```

```python
for column in df.columns:
    mean_value = df[column].mean()
    df[column].fillna(mean_value,inplace=True)

df.isna().sum()
```

```python
X = df.drop('MEDV',axis=1)
y = df['MEDV']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()

model.add(Dense(128, input_shape=(13,) ,activation='relu', name='dense_1')) #Input shape not size
model.add(Dense(64,activation='relu', name='dense_2'))
model.add(Dense(1,activation='linear', name='dense_output'))

model.compile(optimizer='adam', loss = 'mse' , metrics=["mae"]) # here loss not error
model.summary()
```

```python
history = model.fit(
    X_train_scaled,
    y_train,
    epochs = 100,
    validation_split = 0.05,
    verbose=1
)
```

```python
history.history
```

```python
mse_nn, mae_nn = model.evaluate(X_test_scaled, y_test)
```

```python
print('Mean Squared Error on test data', mse_nn)
print('Mean Absolute Error on test data', mae_nn)
```
